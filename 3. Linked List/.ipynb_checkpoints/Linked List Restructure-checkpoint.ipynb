## I. quick sort for linked list
https://leetcode.com/problems/sort-list/ 
1. change the structure, always remember to set:  $ end.next = None$
2. tricky part, set:  pivot.next,less.next,more.next=None,None,Nonepivot.next,less.next,more.next=None,None,None
3. quick sort can have great advantage to sort linked list, need no extra space

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head:
            return None
        if not head.next:
            return head
        #quick sort
        pivot = head
        dummy1 = ListNode(0)
        dummy2 = ListNode(0)
        less, more = dummy1, dummy2
        p = head.next
        while p:
            if p.val == pivot.val:
                pivot.next = p
                pivot = pivot.next
            elif p.val < pivot.val:
                less.next = p
                less = less.next
            elif p.val > pivot.val:
                more.next = p
                more = more.next
            p = p.next
        pivot.next, less.next, more.next = None, None, None
        less = self.sortList(dummy1.next)
        more = self.sortList(dummy2.next)
        if less:
            ret = less
            while less.next:
                less = less.next
            less.next = head
        else:
            ret = head
        while head and head.next:
            head = head.next
        head.next = more
        return ret
```

```python
#82. Remove Duplicates from Sorted List II
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def deleteDuplicates(self, head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    dummy = ListNode(0)
    last, current = dummy, head
    while current and current.next:
        distinct = True
        while current and current.next and current.val == current.next.val:
            current = current.next
            distinct = False
        if distinct:
            last.next = current
            last = last.next
        current = current.next
    # here is very important
    # when you get out of the loop, please remember to change the last node, set last.next = None
    if current:
        last.next = current
    else:
        last.next = None
    return dummy.next     
```

## II Merge Two Sorted Lists
https://leetcode.com/problems/merge-two-sorted-lists/

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if not l1:
            return l2
        if not l2:
            return l1
        # make sure l1 has the smallest val
        if l1.val>l2.val:
            l1, l2 = l2, l1
        dummy = ListNode(0)
        dummy.next = l1
        p = dummy
        while l1 and l2:
            while l1 and l2 and l1.val<=l2.val:
                p = p.next
                l1 = l1.next
            p.next = l2
            while l1 and l2 and l2.val<l1.val:
                p = p.next
                l2 = l2.next
            p.next = l1
        if l1:
            p.next = l1
        if l2:
            p.next = l2
        return dummy.next
```

## III. Merge k Sorted Lists
https://leetcode.com/problems/merge-k-sorted-lists/
1. divide and conquer
2. priority queue

```python
def mergesort(lists):
    if len(lists) == 0:
        return None
    if len(lists) == 1:
        return lists[0]
    p1, p2 = lists[0], lists[1]
    dummy = ListNode(0)
    ret = dummy
    while p1 and p2:
        while p1 and p2 and p1.val<=p2.val:
            ret.next = p1
            p1 = p1.next
            ret = ret.next
        while p1 and p2 and p1.val>=p2.val:
            ret.next = p2
            p2 = p2.next
            ret = ret.next
    if p1:
        ret.next = p1
    if p2:
        ret.next = p2
    return dummy.next


class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        if len(lists) <= 2:
            return mergesort(lists)
        half1 = self.mergeKLists(lists[:len(lists)//2])
        half2 = self.mergeKLists(lists[len(lists)//2:])
        return mergesort([half1, half2])
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        dummy = ListNode(0)
        p = dummy
        import heapq 
        l = [(lists[i].val, lists[i]) for i in range(len(lists)) if lists[i]]
        heapq.heapify(l)
        while l:
            p.next = heapq.heappop(l)[1]
            p = p.next
            if p.next:
                heapq.heappush(l,(p.next.val, p.next))
        return dummy.next
```
